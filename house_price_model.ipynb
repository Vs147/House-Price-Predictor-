```python
def bhk_outlier_handle(df):
    convert  = np.array([])  # Initialize an empty NumPy array to store indices of rows to be dropped

    # Grouping the DataFrame by location
    for location, location_df in df.groupby('location'):
        bhk_brief = {}  # Dictionary to store statistics for each BHK (Bedrooms, Hall, Kitchen) configuration

        # Grouping the location DataFrame by the number of bedrooms (bhk)
        for bhk, bhk_df in location_df.groupby('bhk'):
#             print(bhk_df)
#             print(bhk)
            # Calculating mean, standard deviation, and count of prices per square foot for each BHK configuration
            bhk_brief[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }
#             print(bhk_brief)
#         Iterating over each BHK configuration again
        for bhk, bhk_df in location_df.groupby('bhk'):
            # Checking if statistics for (bhk-1) exists and the count is greater than 5
            stats = bhk_brief.get(bhk - 1)
#             print(stats)
            if stats and stats['count'] > 5:
                # Finding indices of rows where price per sqft is less than the mean of (bhk-1)
                convert = np.append(convert, bhk_df[bhk_df.price_per_sqft < (stats['mean'])].index.values)
#                 print(bhk_df)

#     Dropping rows with outlier prices
    return df.drop(convert, axis='index')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.preprocessing import StandardScaler,OneHotEncoder
from sklearn.compose import make_column_transformer

from sklearn.pipeline import make_pipeline

from sklearn.metrics import r2_score
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 0)
print(X_train.shape)
print(X_test.shape)
```

```python
column_trans = make_column_transformer((OneHotEncoder(sparse = False),['location']),remainder = 'passthrough')
```

```python
column_trans
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()

```

```python
pipe = make_pipeline(column_trans,scaler,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```
